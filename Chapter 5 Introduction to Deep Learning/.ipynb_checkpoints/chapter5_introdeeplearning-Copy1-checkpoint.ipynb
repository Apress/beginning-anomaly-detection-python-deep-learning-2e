```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
x, y = make_moons(n_samples=5000, noise=0.1, random_state=42)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MSE
```

```python

## Model 1

inp = Input(shape=(x_train.shape[-1],))
l1 = Dense(3, activation='relu', kernel_initializer='he_normal')(inp)
l2 = Dense(3, activation='relu', kernel_initializer='he_normal')(l1)
l2 = Dense(3, activation='relu', kernel_initializer='he_normal')(l2)
l2 = Dense(3, activation='relu', kernel_initializer='he_normal')(l2)
l3 = Dense(1, activation=None)(l2)

m1 = Model(inp, l3)
```

```python
m1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
epochs = 50

m1.fit(x_train, y_train, batch_size=128, epochs=epochs)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
def evaluate(y_true, y_pred):
    f1 = f1_score(y_true, y_pred)
    p = precision_score(y_true, y_pred)
    r = recall_score(y_true, y_pred)    
    
    print(f'Precision: {p}')
    print(f'Recall: {r}')    
    print(f'F1-Score: {f1}')    
```

```python
preds = m1.predict(x_test)
```

```python
preds = np.where(preds < 0.5, 0, 1)
```

```python
evaluate(y_test, preds)
```

```python
true_fn = lambda x: 0.9*x**3 - 0.5*x**2 + 0.07*x 
```

```python
x = np.linspace(-1, 1, num=10000)
noise = np.random.normal(loc=0, scale=0.03, size=len(x))
y_true = true_fn(x)
y = y_true + noise
```

```python
plt.scatter(x, y)
plt.plot(x, y_true, c='red')
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
indices = range(len(y))
train_len = int(0.8 * len(y))
train_indices = np.random.choice(indices, size=train_len, replace=False)
test_indices = list(set(indices).difference(set(train_indices)))
x_train = x[train_indices]
x_test = x[test_indices]

y_train = y[train_indices]
y_test = y[test_indices]
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
# model

inp = Input(shape=(x_train.shape[-1], ))
l1 = Dense(3, activation='relu')(inp)
l2 = Dense(3, activation='relu')(l1)
l3 = Dense(3, activation='relu')(l2)
l4 = Dense(3, activation='relu')(l3)
out = Dense(1, activation=None)(l4)

m = Model(inp, out)
m.summary()
```

```python
m.compile(optimizer='adam', loss='mean_squared_error', metrics='mean_absolute_error')
```

```python
epochs = 50
m.fit(x_train, y_train, batch_size=256, epochs=epochs)
```

```python
preds = m.predict(x_test)
```

```python
# plt.scatter(range(len(preds)), preds)
# plt.scatter(test_indices, y_test)
plt.scatter(test_indices, preds)
# plt.scatter(train_indices, y_train, c='yellow')
plt.plot(y_true, c='red')
```
