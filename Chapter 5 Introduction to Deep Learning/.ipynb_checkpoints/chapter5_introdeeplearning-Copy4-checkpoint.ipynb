```python
import numpy as np
true_fn = lambda x: 0.9*x**3 - 0.5*x**2 + 0.07*x 
```

```python
x = np.linspace(-1, 1, num=10000)
noise = np.random.normal(loc=0, scale=0.03, size=len(x))
y_true = true_fn(x)
y = y_true + noise
```

```python
plt.scatter(x, y)
plt.plot(x, y_true, c='red')
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
indices = range(len(y))
train_len = int(0.8 * len(y))
train_indices = np.random.choice(indices, size=train_len, replace=False)
test_indices = list(set(indices).difference(set(train_indices)))
x_train = x[train_indices]
x_test = x[test_indices]

y_train = y[train_indices]
y_test = y[test_indices]
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
plt.title("Train Values")
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x_train, y_train)
```

```python
# Manually Computed Forward Pass 
w1 = np.array([[-0.1, 0.9, 0.3]])
b1 = np.array([0.1, 0.1, -0.05])
```

```python
z1 = x_train @ w1 + b1
a1 = np.maximum(0, z1)
```

```python
plt.title("Layer 1 ReLU Output")
plt.xlabel("X Index Value")
plt.ylabel("Activation")

plt.scatter(train_indices, z1[:, 0], c='pink')
plt.scatter(train_indices, z1[:, 1], c='lightgreen')
plt.scatter(train_indices, z1[:, 2], c='skyblue')

plt.scatter(train_indices, a1[:, 0], c='red')
plt.scatter(train_indices, a1[:, 1], c='green')
plt.scatter(train_indices, a1[:, 2], c='blue')
```

```python
w2 = np.array([[2.3, -0.9, 1.2 ], [-0.9, 0.92, -0.2 ], [5.3, -0.2, 0.1 ]])
b2 = np.array([-0.1, -0.1, 0.1])
```

```python
z2 = a1 @ w2 + b2
a2 = np.maximum(0, z2)
```

```python
plt.title("Layer 2 ReLU Output")
plt.xlabel("X Index Value")
plt.ylabel("Activation")

plt.scatter(train_indices, z2[:, 0], c='pink')
plt.scatter(train_indices, z2[:, 1], c='lightgreen')
plt.scatter(train_indices, z2[:, 2], c='skyblue')

plt.scatter(train_indices, a2[:, 0], c='red')
plt.scatter(train_indices, a2[:, 1], c='green')
plt.scatter(train_indices, a2[:, 2], c='blue')
```

```python
w3 = np.array([[1.0, 0.9, -1.1 ], [1.1, 0.92, 0.5 ], [-0.1, -0.9, -0.1 ]])
b3 = np.array([0.2, -0.4, 0.1])

z3 = a2 @ w3 + b3
a3 = np.maximum(0, z3)
```

```python
plt.title("Layer 3 ReLU Output")
plt.xlabel("X Index Value")
plt.ylabel("Activation")

plt.scatter(train_indices, z3[:, 0], c='pink')
plt.scatter(train_indices, z3[:, 1], c='lightgreen')
plt.scatter(train_indices, z3[:, 2], c='skyblue')

plt.scatter(train_indices, a3[:, 0], c='red')
plt.scatter(train_indices, a3[:, 1], c='green')
plt.scatter(train_indices, a3[:, 2], c='blue')
```

```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model

from sklearn.metrics import mean_squared_error
```

```python
# model

inp = Input(shape=(x_train.shape[-1], ))
l1 = Dense(3, activation='tanh')(inp)
l2 = Dense(3, activation='tanh')(l1)
l3 = Dense(3, activation='tanh')(l2)
l4 = Dense(3, activation='tanh')(l3)
out = Dense(1, activation=None)(l4)

m = Model(inp, out)
m.summary()
```

```python
m.compile(optimizer='adam', loss='mean_squared_error')
```

```python
epochs = 50

m.fit(x_train, y_train, batch_size=128, epochs=epochs)
```

```python
preds = m.predict(x_test)
loss = mean_squared_error(true_fn(x_test), preds)

plt.suptitle("Neural Network using Tanh")
plt.title(f"Loss: {loss:.6f}", fontsize=8)
plt.scatter(test_indices, preds, label='Approximation')
plt.plot(y_true, c='red', label='True Function')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error

# true:
fn1 = lambda x: 4*x + 3

# randomly intialize m and b
m = 1
b = 0
```

```python
# Take four points to optimize m and b on
x = np.array([0, 1, 2, 3])
y = np.array([fn1(f) for f in x])

x, y
```

```python
preds = np.array([m*f + b for f in x])
preds
```

```python
m_grid = np.linspace(0, 5)
b_grid = np.linspace(0, 5)

mm, bb = np.meshgrid(m_grid, b_grid)

pairs = np.asarray(list(zip(mm.reshape(-1), bb.reshape(-1))))

mses = []
for f in range(len(pairs)):
    m, b = pairs[f]
    preds = [g*m + b for g in x]
    
    mse_val = mean_squared_error(y, preds)

    mses.append(mse_val)
mses = np.array(mses)
```

```python

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(projection='3d')

ax.view_init(elev=10, azim=-55)
ax.set_box_aspect(aspect=None, zoom=0.8)

ax.scatter(pairs[:, 0], pairs[:, 1], mses.reshape(-1, 1), alpha=0.3)

ax.scatter([4], [3], [0], s=200, color='red')
ax.text(4, 3, -15, 'Optimal')

m = 1
b = 0

preds = [g*m + b for g in x]
mse_val = mean_squared_error(y, preds)
ax.scatter([m], [b], [mse_val], s=100, color='red')
ax.text(0.4, 0, 55, 'Initial')

ax.set_xlabel('m')
ax.set_ylabel('b')
ax.set_zlabel('Loss')

```

```python
def mse_grad(y_true, y_pred):
    # We will perform the averaging in a later step instead of right here
    grad = -2 * (y_true-y_pred)
    return grad
```

```python
# Initial settings

m = 1 
b = 0

preds = [m * g + b for g in x]

plt.title(f"Estimated Line for m={m} and b={b}")
plt.plot(x, y)
plt.plot(x, preds)
plt.legend(["True Line", "Predicted Line"])
```

```python
m = 1
b = 0

step_size = 0.001
for f in range(1000):
    preds = np.array([m*g + b for g in x])
    loss = mean_squared_error(y,  preds)
    print(f'\r{f}: {loss}', end='')
    
    dL_dyhat = mse_grad(y, preds)
    dyhat_dm = x
    dyhat_db = np.ones(len(x))

    dL_dm = np.mean(dL_dyhat * dyhat_dm)
    dL_db = np.mean(dL_dyhat * dyhat_db)
    
    m = m - step_size * dL_dm
    b = b - step_size * dL_db
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Input

input_layer = Input(shape=(28,28, 1))
layer1 = Conv2D(kernel_size=3, filters=32, strides=2, activation='relu')(input_layer)
layer2 = Conv2D(kernel_size=3, filters=16, strides=2, activation='relu')(layer1)
layer3 = Conv2D(kernel_size=3, filters=8, strides=2, activation='relu')(layer2)
flat = Flatten()(layer3)
output_layer = Dense(10, activation='softmax')(flat)

model = Model(input_layer, output_layer)
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=128, validation_data=(x_val, y_val), epochs=64)
```

```python
history_dict = history.history
```

```python
plt.title("Loss Curve")

plt.plot(history_dict['loss'], label='training_loss')
plt.plot(history_dict['val_loss'], label='validation_loss')

plt.vlines(np.argmin(history_dict['val_loss']), 0, np.max(history_dict['loss']),
           label = 'overfitting', linestyles='dashed', color='red')

plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
```

```python
print("Overfitting Boundary Epoch:", np.argmin(history_dict['val_loss']))
```

```python
plt.title("Accuracy Curve")

plt.plot(history_dict['accuracy'], label='training_accuracy')
plt.plot(history_dict['val_accuracy'], label='validation_accuracy')

plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
```

```python
history = model.evaluate(x_test, y_test)
```
